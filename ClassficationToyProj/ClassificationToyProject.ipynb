```python
#Step1
x=df.iloc[:,0:2]
x.shape
x
```

```python
y=df.iloc[:,-1]
y.shape
y
```

```python
!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1) #inp,out,fraction of data to hide
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

```python
#Train the model
#We have decided to use Logistic Regression so import that
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
#Train the object clf of LogisticRegression
clf.fit(x_train,y_train)
clf
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#accuracy=80%
```

```python
#What is the decision boundary that our model has drawn , to see that we will use:
# !pip install mlxtend
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train.values,clf=clf,legend=2)
```
